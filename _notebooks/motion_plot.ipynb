```python
import numpy as np 
import matplotlib.pyplot as plt 
 
values = np.random.uniform(-1, 1, 1000000) 
count, bins, ignored = plt.hist(values, 10000, density=True, color = 'grey')
plt.axvline(x = 0.3, color = 'black', ls=':', lw=2, label = 'ramdom motion speed - user specified')
plt.axvline(x = -0.3, color = 'black', ls=':', lw=2)

plt.legend(bbox_to_anchor = (1.0, 1.2))
plt.title('Uniform Distribution')
plt.ylabel('Density')
plt.xlabel('Values')
plt.show()
```
